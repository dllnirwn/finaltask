```python
# Pre Processing
import pandas as pd
import numpy as np
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1)) #inisialisasi normalisasi MinMax
data = min_max_scaler.fit_transform(df3) #transformasi MinMax untuk fitur
```

```python
target.value_counts()
```

```python
dataset = pd.DataFrame({'FLAG_OWN_CAR':data[:,0],'FLAG_OWN_REALTY':data[:,1],'AMT_INCOME_TOTAL':data[:,2],'AMT_CREDIT':data[:,3],'AMT_GOODS_PRICE':data[:,4],'AMT_ANNUITY':data[:,5],'NAME_INCOME_TYPE':data[:,6],'NAME_HOUSING_TYPE':data[:,7]})
dataset
```

```python
from sklearn.linear_model import LogisticRegression 

from sklearn.model_selection import train_test_split 

X_train,X_test,y_train,y_test=train_test_split(dataset,target.TARGET,test_size=0.25,random_state=0) 
```

```python
# Deklarasi Model
logreg =  LogisticRegression(solver='liblinear') 

# fit model dengan data

logreg.fit(X_train,y_train) 

# Membuat Prediksi

y_pred=logreg.predict(X_test) 

y_pred

from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```
